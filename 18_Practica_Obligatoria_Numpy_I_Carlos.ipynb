Por ejemplo, si Rafa compra en *Manolo Bakes* gastaría:  

$6\cdot1.50+5\cdot2+3\cdot5+1\cdot16=50$  

y si compra en *Starbucks* gastaría:  

$6\cdot1+5\cdot2.50+3\cdot4.50+1\cdot17=49$  

**SE PIDE**: Genera la matriz que exprese la cantidad que cada persona gastaría en cada tienda, y recomienda a cada uno dónde debería hacer sus compras (pista: donde le cueste menos, ¿no?)

**REQUISITIO**: Emplea Numpy, su tipo array y sus operaciones para la resolución del ejercicio. 

```python
import numpy as np
```

```python
mat_bollos = np.array([
    [6, 5, 3, 1],
    [3, 6, 2, 2],
    [3, 4, 3, 1]
])

print(mat_bollos)
```

```python
# Precios por unidad en cada cafetería
tarifas_manolo = np.array([1.5, 2.0, 5.0, 16.0])
tarifas_starbucks = np.array([1.0, 2.5, 4.5, 17.0])

# Matriz 2x4 con los precios de ambas cafeterías
mat_precios = np.array([tarifas_manolo, tarifas_starbucks])

print(mat_precios)

```

```python
# Transponemos para que las dimensiones cuadren en el producto matricial
mat_precios = mat_precios.T  # ahora es 4x2

print(mat_precios)
```

```python
# Matriz coste: cuánto paga cada persona en cada cafetería
mat_coste = np.dot(mat_bollos,mat_precios)

print(mat_coste)
```

```python
cafeterias = ["Manolo Bakes", "Starbucks"]
clientes = ["Rafa", "Nacho", "Clara"]

for indice_cliente, nombre_cliente in enumerate(clientes):
    # Comparamos el coste en cada cafetería para ese cliente
    if mat_coste[indice_cliente][0] != mat_coste[indice_cliente][1]:
        idx_mejor_opcion = np.argmin(mat_coste[indice_cliente])
        mejor_local = cafeterias[idx_mejor_opcion]
        print(f"Para {nombre_cliente} la opción más barata es {mejor_local}")
    else:
        print(f"Para {nombre_cliente} el coste es el mismo en ambas cafeterías")

        
```

II.1: Similitud entre películas

Vamos a calcular similitudes entre películas. Para ello las clasificasmos según generos, en concreto tenemos:

* Toy Story: Aventura, Animación, Comedia
* Jumanji: Aventura   
* Mortal Kombat: Acción, Aventura  
* Pocahontas: Animación, Romance

Y ahora calcularemos su similitud por el método del coseno. Para ello:

1. Convierte cada película en un vector y almacenalo en un diccionario con clave el nombre de la película. Pista: Para convertirlo en vector, crea un vector/lista por película con esta estructura (¿qué valores puedes darle por género?):

$$[Acción,  Aventura , Animación, Comedia, Romance]$$


```python
catalogo_peliculas = {
    "Toy Story":      [0, 1, 1, 1, 0],
    "Jumanji":        [0, 1, 0, 0, 0],
    "Mortal Kombat":  [1, 1, 0, 0, 0],
    "Pocahontas":     [0, 0, 1, 0, 1]
}
```



2. Crea una matriz de películas con un array de numpy. Muestra la dimension, shape, tamaño y tipo del array.


```python
vectores_valoracion = list(catalogo_peliculas.values())
matriz_valoraciones = np.array(vectores_valoracion)

print(matriz_valoraciones)
```

```python
# Recomendación de película similar para cada una (sin Wall-E aún)
for pelicula_ref, vector_ref in catalogo_peliculas.items():
    mejor_similar = None
    sim_max = -1.0

    for pelicula_cand, vector_cand in catalogo_peliculas.items():
        if pelicula_ref == pelicula_cand:
            continue

        numerador = np.dot(vector_ref, vector_cand)
        denom = np.linalg.norm(vector_ref) * np.linalg.norm(vector_cand)
        coseno_sim = numerador / denom

        if coseno_sim > sim_max:
            sim_max = coseno_sim
            mejor_similar = pelicula_cand

    print(f"Para {pelicula_ref} la más similar es {mejor_similar}, similitud: {sim_max:.4f}")
```
